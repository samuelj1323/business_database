```python
import csv
from csv import reader
csv.field_size_limit(1000000000)


with open("./Insertion/locations.csv", 'w', encoding="utf-8") as loc_file:
    loc_file.write("Loc_Address,Loc_City,Loc_State,Loc_Zip,Loc_Latitude,Loc_Longitude\n")
    with open("./Insertion/categories.csv", 'w', encoding="utf-8") as cat_file:
        cat_file.write("Cat_Name\n")
        with open("./Insertion/businesses.csv", 'w', encoding="utf-8") as bus_file:
            bus_file.write("Bus_ID,Loc_ID,Cat_ID,Bus_Name,Bus_Rating,Bus_ReviewCount\n")
            with open("yelp_academic_dataset_business.csv", 'r', encoding='utf-8') as bus_csv:
                bus_read = list(reader(bus_csv, delimiter=','))

                cats = {}
                first = True
                i = 0
                offset = 0
                for row in bus_read:
                    if first:
                        first = False
                        continue

                    index = 0
                    address = ""
                    while row[2].find("'" , index) != -1:
                        qin = row[2].find("'" , index)
                        address += row[2][index:qin] + "''"
                        index = qin+1
                    address += row[2][index:]
                    
                    index = 0
                    while address.find("\"", index) != -1:
                        qin = address.find("\"", index)
                        temp = address[0:qin] + "''"
                        address = temp + address[qin+1:]
                        index = qin+2

                    index = 0
                    city = ""
                    while row[3].find("'" , index) != -1:
                        qin = row[3].find("'" , index)
                        city += row[3][index:qin] + "''"
                        index = qin+1
                    city += row[3][index:]
                    
                    index = 0
                    while city.find("\"", index) != -1:
                        qin = city.find("\"", index)
                        temp = city[0:qin] + "''"
                        city = temp + city[qin+1:]
                        index = qin+2

                    loc_file.write("\"" + address + "\",\"" + city + "\"," + row[4] + "," + row[5] + "," + row[6] + "," + row[7] + "\n");

                    cat_ID = 2000000 + i - offset
                    if row[11] in cats:
                        cat_ID = cats[row[11]]
                        offset += 1
                    else:
                        cats[row[11]] = cat_ID

                        index = 0
                        text = ""
                        while row[11].find("'" , index) != -1:
                            qin = row[11].find("'" , index)
                            text += row[11][index:qin] + "''"
                            index = qin+1
                        text += row[11][index:]
                    
                        index = 0
                        while text.find("\"", index) != -1:
                            qin = text.find("\"", index)
                            temp = text[0:qin] + "''"
                            text = temp + text[qin+1:]
                            index = qin+2

                        cat_file.write("\"" + text + "\"\n");

                    index = 0
                    text = ""
                    while row[1].find("'" , index) != -1:
                        qin = row[1].find("'" , index)
                        text += row[1][index:qin] + "''"
                        index = qin+1
                    text += row[1][index:]
                    
                    index = 0
                    while text.find("\"", index) != -1:
                        qin = text.find("\"", index)
                        temp = text[0:qin] + "''"
                        text = temp + text[qin+1:]
                        index = qin+2

                    bus_file.write(row[0] + "," + str(1000000+i) + "," + str(cat_ID) + ",\"" + text + "\"," + row[8] + "," + row[9] + "\n");

                    i += 1
                print(i)

with open("./Insertion/users.csv", 'w', encoding='utf-8') as usr_file:
    usr_file.write("Usr_ID,Usr_FName,Usr_ReviewCount,Usr_JoinDate,Usr_AverageRating,Usr_CurrentlyElite,Usr_ComplimentsRecv\n")
    with open("./Insertion/elitehistory.csv", 'w', encoding='utf-8') as hist_file:
        hist_file.write("Usr_ID,Hist_Year,Hist_Current\n")
        with open("yelp_academic_dataset_user.csv", 'r', encoding='utf-8') as usr_csv:
            usr_read = list(reader(usr_csv, delimiter=','))

            first = True
            i = 0
            for row in usr_read:
                if first:
                    first = False
                    continue

                index = 0
                text = ""
                while row[1].find("'" , index) != -1:
                    qin = row[1].find("'" , index)
                    text += row[1][index:qin] + "''"
                    index = qin+1
                text += row[1][index:]
                    
                index = 0
                while text.find("\"", index) != -1:
                    qin = text.find("\"", index)
                    temp = text[0:qin] + "''"
                    text = temp + text[qin+1:]
                    index = qin+2

                cur_elite = (1 if (row[7].find("2020") != -1) else 0)
                compliments = int(row[4]) + int(row[5]) + int(row[6])

                usr_file.write(row[0] + ",\"" + text + "\"," + row[2] + "," + row[3] + "," + row[10] + "," + str(cur_elite) + "," + str(compliments) + "\n")

                cur_elite = 0
                for elite_year in row[7].split(","):
                    if elite_year == "":
                        continue
                    if elite_year == 2020:
                        cur_elite = 1
                    hist_file.write(row[0] + "," + (elite_year+"-01-01") + "," + str(cur_elite) + "\n")

                i += 1
            print(i)

with open("./Insertion/reviews.csv", 'w', encoding='utf-8') as rev_file:
    rev_file.write("Rev_ID,Usr_ID,Bus_ID,Rev_Rating,Rev_Date,Rev_Text,Rev_ComplimentsRecv\n")
    with open("yelp_academic_dataset_review.csv", 'r', encoding='utf-8') as rev_csv:
        rev_read = list(reader(rev_csv, delimiter=','))

        first = True
        i = 0
        for row in rev_read:
            if first:
                first = False
                continue

            compliments = int(row[4]) + int(row[5]) + int(row[6])

            index = 0
            text = ""
            while row[7].find("'" , index) != -1:
                qin = row[7].find("'" , index)
                text += row[7][index:qin] + "''"
                index = qin+1
            text += row[7][index:]
                    
            index = 0
            while text.find("\"", index) != -1:
                qin = text.find("\"", index)
                temp = text[0:qin] + "''"
                text = temp + text[qin+1:]
                index = qin+2

            rating = int(float(row[3]))
            rev_file.write(row[0] + "," + row[1] + "," + row[2] + "," + str(rating) + "," + row[8] + ",\"" + text + "\"," + str(compliments) + "\n")

            i += 1
        print(i) 

with open("./Insertion/tips.csv", 'w', encoding='utf-8') as tip_file:
    tip_file.write("Usr_ID,Bus_ID,Tip_Date,Tip_Text,Tip_ComplimentsRecv\n")
    with open("yelp_academic_dataset_tip.csv", 'r', encoding='utf-8') as tip_csv:
        tip_read = list(reader(tip_csv, delimiter=','))

        first = True
        i = 0
        for row in tip_read:
            if first:
                first = False
                continue

            index = 0
            text = ""
            while row[2].find("'" , index) != -1:
                qin = row[2].find("'" , index)
                text += row[2][index:qin] + "''"
                index = qin+1
            text += row[2][index:]
                    
            index = 0
            while text.find("\"", index) != -1:
                qin = text.find("\"", index)
                temp = text[0:qin] + "''"
                text = temp + text[qin+1:]
                index = qin+2

            tip_file.write(row[0] + "," + row[1] + "," + row[3] + ",\"" + text + "\"," + row[4] + "\n")

            i += 1
        print(i)
```
